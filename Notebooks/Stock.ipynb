```python
# ── Tableau complet des valeurs manquantes
missing_count = df.isnull().sum()
missing_pct = (df.isnull().sum() /  len(df) * 100).round(2)
total_missing = df.isnull().sum().sum()

missing_df = pd.DataFrame({
    'Valeurs manquantes' : missing_count,
    'Pourcentage (%)' : missing_pct,
    'Statut' : ['Complet' if v == 0 else 'Manquant' for v in missing_count]
})
print('── Analyse des Valeurs Manquantes ──')
display(missing_df)
print()
print(f'Total valeurs manquantes : {total_missing}')
print(f'Le dataset est COMPLET — aucun traitement d\'imputation nécessaire')
```

```python
# ── Visualisation heatmap des valeurs manquantes
fig, axes = plt.subplots(1,2, figsize=(16,6))

# -- Heatmap
sns.heatmap(
    df.isnull(),
    ax=axes[0],
    cbar=False,
    cmap='viridis',
    yticklabels=False
)
axes[0].set_title('Heatmap des valeurs manquantes\n(jaune = manquant, violet = présent)', fontsize=13)
axes[0].set_xlabel('Colonnes')
axes[0].tick_params(axis='x', rotation=90)

# -- Barplot du pourcentage de complétude
completude = (1 - df.isnull().mean()) * 100
colors = ['#2ecc71' if v == 100 else '#e74c3c' for v in completude]
axes[1].barh(completude.index, completude.values, color=colors, edgecolor='white')
axes[1].set_xlim(0, 110)
axes[1].axvline(100, color='black', linestyle='--', linewidth=1, alpha=0.5)
axes[1].set_title('Complétude des colonnes (%)\n(vert = 100% complet)', fontsize=13)
axes[1].set_xlabel('Complétude (%)')
for i, v in enumerate(completude.values):
    axes[1].text(v + 0.5, i, f'{v:.1f}%', va='center', fontsize=8)

plt.suptitle('Analyse des Valeurs Manquantes', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('phase1_missing_values.png', bbox_inches='tight', dpi=150)
plt.show()
print('Résultat : Aucune valeurs manquantes — Dataset 100% complet')
```

```python
# ── Distribution des variables numériques clés : histogramme + KDE + boxplot
num_cols = ['stock_initial', 'achat_qte', 'vente_qte', 'stock_final',
            'prix_achat', 'prix_vente', 'rotation_stock', 'couverture_stock',
            'moyenne_vente_7j', 'moyenne_vente_30j', 'ecart_stock']

fig, axes = plt.subplots(len(num_cols), 2, figsize=(16, 5 * len(num_cols)))

for i, col in enumerate(num_cols):
    data = df[col].dropna()

    # -- Histogramme + KDE
    axes[i, 0].hist(data, bins=50, color='steelblue', alpha=0.7, edgecolor='white', density=True)
    data.plot.kde(ax=axes[i, 0], color='darkblue', linewidth=2)
    axes[i, 0].set_title(f'Distribution de {col}', fontsize=12)
    axes[i, 0].set_xlabel(col)
    axes[i, 0].set_ylabel('Densité')
    axes[i, 0].axvline(data.mean(),   color='red',    linestyle='--', linewidth=1.5, label=f'Moyenne : {data.mean():.2f}')
    axes[i, 0].axvline(data.median(), color='orange', linestyle='--', linewidth=1.5, label=f'Médiane : {data.median():.2f}')
    axes[i, 0].legend(fontsize=9)

    # -- Boxplot
    axes[i, 1].boxplot(data, vert=True, patch_artist=True,
                        boxprops=dict(facecolor='lightblue', color='navy'),
                        medianprops=dict(color='red', linewidth=2),
                        flierprops=dict(marker='o', color='red', alpha=0.3, markersize=3))
    axes[i, 1].set_title(f'Boxplot de {col}', fontsize=12)
    axes[i, 1].set_ylabel(col)

    # Statistiques dans le titre
    skew = data.skew()
    kurt = data.kurtosis()
    axes[i, 0].set_xlabel(f'{col}  |  Skewness={skew:.2f}  Kurtosis={kurt:.2f}')

plt.suptitle('Analyse Univariée — Variables Numériques', fontsize=16, fontweight='bold', y=1.002)
plt.tight_layout()
plt.savefig('phase1_univariate_numeric.png', bbox_inches='tight', dpi=120)
plt.show()
print('Graphiques enregistrés : phase1_univariate_numeric.png')
```

```python
# ── Analyse spécifique de la variable cible : vente_qte
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Distribution
axes[0].hist(df['vente_qte'], bins=30, color='#3498db', edgecolor='white', alpha=0.8)
axes[0].set_title('Distribution des Ventes (vente_qte)', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Quantité vendue par jour')
axes[0].set_ylabel('Fréquence')
axes[0].axvline(df['vente_qte'].mean(),   color='red',    linestyle='--', linewidth=2, label=f'Moy: {df["vente_qte"].mean():.1f}')
axes[0].axvline(df['vente_qte'].median(), color='orange', linestyle='--', linewidth=2, label=f'Med: {df["vente_qte"].median():.1f}')
axes[0].legend()

# Ventes par jour de semaine
jours = ['Lundi','Mardi','Mercredi','Jeudi','Vendredi','Samedi','Dimanche']
ventes_jour = df.groupby('jour_semaine')['vente_qte'].mean().reindex(range(7))
axes[1].bar(jours, ventes_jour.values, color='#2ecc71', edgecolor='white')
axes[1].set_title('Ventes moyennes par jour de semaine', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Jour de la semaine')
axes[1].set_ylabel('Ventes moyennes')
axes[1].tick_params(axis='x', rotation=30)

# Ventes par mois
ventes_mois = df.groupby('mois')['vente_qte'].mean()
mois_labels = ['Jan','Fév','Mar','Avr','Mai','Jun','Jul','Aoû','Sep','Oct','Nov','Déc']
axes[2].bar(ventes_mois.index, ventes_mois.values, color='#e67e22', edgecolor='white')
axes[2].set_title('Ventes moyennes par mois', fontsize=13, fontweight='bold')
axes[2].set_xlabel('Mois')
axes[2].set_ylabel('Ventes moyennes')
axes[2].set_xticks(ventes_mois.index)
axes[2].set_xticklabels(mois_labels[:len(ventes_mois)], rotation=30)

plt.suptitle('Variable Cible : Quantité Vendue (vente_qte)', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('phase1_target_variable.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
# ── Distribution des variables catégorielles
cat_vars = {
    'categorie' : 'Catégorie de bijou',
    'matiere'   : 'Matière du bijou',
    'ville'     : 'Ville / Dépôt',
    'is_weekend': 'Jour weekend (0=Non, 1=Oui)'
}

fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

idx = 0
for col, title in cat_vars.items():
    counts = df[col].value_counts()

    # -- Barplot
    axes[idx].bar(counts.index.astype(str), counts.values,
                  color=sns.color_palette('Set2', len(counts)), edgecolor='white')
    axes[idx].set_title(f'{title}\n(Barplot)', fontsize=11, fontweight='bold')
    axes[idx].set_ylabel('Nombre de lignes')
    axes[idx].tick_params(axis='x', rotation=20)
    for j, v in enumerate(counts.values):
        axes[idx].text(j, v + 100, f'{v:,}', ha='center', fontsize=8)

    # -- Pie chart
    axes[idx + 1].pie(counts.values, labels=counts.index.astype(str),
                      autopct='%1.1f%%', startangle=90,
                      colors=sns.color_palette('Set2', len(counts)))
    axes[idx + 1].set_title(f'{title}\n(Répartition %)', fontsize=11, fontweight='bold')
    idx += 2

plt.suptitle('Analyse Univariée — Variables Catégorielles', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('phase1_univariate_cat.png', bbox_inches='tight', dpi=150)
plt.show()
```
