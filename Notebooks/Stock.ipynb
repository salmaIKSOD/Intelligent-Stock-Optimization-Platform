         6 - Analyse des valeurs manquantes :  
    </p>
</div>

```python
# ── Tableau complet des valeurs manquantes
missing_count = df.isnull().sum()
missing_pct = (df.isnull().sum() /  len(df) * 100).round(2)
total_missing = df.isnull().sum().sum()

missing_df = pd.DataFrame({
    'Valeurs manquantes' : missing_count,
    'Pourcentage (%)' : missing_pct,
    'Statut' : ['Complet' if v == 0 else 'Manquant' for v in missing_count]
})
print('── Analyse des Valeurs Manquantes ──')
display(missing_df)
print()
print(f'Total valeurs manquantes : {total_missing}')
print(f'Le dataset est COMPLET — aucun traitement d\'imputation nécessaire')
```

```python
# ── Visualisation heatmap des valeurs manquantes
fig, axes = plt.subplots(1,2, figsize=(16,6))

# -- Heatmap
sns.heatmap(
    df.isnull(),
    ax=axes[0],
    cbar=False,
    cmap='viridis',
    yticklabels=False
)
axes[0].set_title('Heatmap des valeurs manquantes\n(jaune = manquant, violet = présent)', fontsize=13)
axes[0].set_xlabel('Colonnes')
axes[0].tick_params(axis='x', rotation=90)

# -- Barplot du pourcentage de complétude
completude = (1 - df.isnull().mean()) * 100
colors = ['#2ecc71' if v == 100 else '#e74c3c' for v in completude]
axes[1].barh(completude.index, completude.values, color=colors, edgecolor='white')
axes[1].set_xlim(0, 110)
axes[1].axvline(100, color='black', linestyle='--', linewidth=1, alpha=0.5)
axes[1].set_title('Complétude des colonnes (%)\n(vert = 100% complet)', fontsize=13)
axes[1].set_xlabel('Complétude (%)')
for i, v in enumerate(completude.values):
    axes[1].text(v + 0.5, i, f'{v:.1f}%', va='center', fontsize=8)

plt.suptitle('Analyse des Valeurs Manquantes', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('phase1_missing_values.png', bbox_inches='tight', dpi=150)
plt.show()
print('Résultat : Aucune valeurs manquantes — Dataset 100% complet')
```
