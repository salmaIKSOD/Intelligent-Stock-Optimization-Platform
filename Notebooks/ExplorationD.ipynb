<h3 style="color: #F39C12; font-weight: bold; font-family: 'Times New Roman';text-align: center;"> Distribution des variables numériques clés  :</h3>


```python
# ── Distribution des variables numériques clés : histogramme + KDE + boxplot
num_cols = ['stock_initial', 'achat_qte', 'vente_qte', 'stock_final',
            'prix_achat', 'prix_vente', 'rotation_stock', 'couverture_stock',
            'moyenne_vente_7j', 'moyenne_vente_30j', 'ecart_stock']

fig, axes = plt.subplots(len(num_cols), 2, figsize=(16, 5 * len(num_cols)))

for i, col in enumerate(num_cols):
    data = df[col].dropna()

    # -- Histogramme + KDE
    axes[i, 0].hist(data, bins=50, color='steelblue', alpha=0.7, edgecolor='white', density=True)
    data.plot.kde(ax=axes[i, 0], color='darkblue', linewidth=2)
    axes[i, 0].set_title(f'Distribution de {col}', fontsize=12)
    axes[i, 0].set_xlabel(col)
    axes[i, 0].set_ylabel('Densité')
    axes[i, 0].axvline(data.mean(),   color='red',    linestyle='--', linewidth=1.5, label=f'Moyenne : {data.mean():.2f}')
    axes[i, 0].axvline(data.median(), color='orange', linestyle='--', linewidth=1.5, label=f'Médiane : {data.median():.2f}')
    axes[i, 0].legend(fontsize=9)

    # -- Boxplot
    axes[i, 1].boxplot(data, vert=True, patch_artist=True,
                        boxprops=dict(facecolor='lightblue', color='navy'),
                        medianprops=dict(color='red', linewidth=2),
                        flierprops=dict(marker='o', color='red', alpha=0.3, markersize=3))
    axes[i, 1].set_title(f'Boxplot de {col}', fontsize=12)
    axes[i, 1].set_ylabel(col)

    # Statistiques dans le titre
    skew = data.skew()
    kurt = data.kurtosis()
    axes[i, 0].set_xlabel(f'{col}  |  Skewness={skew:.2f}  Kurtosis={kurt:.2f}')

plt.suptitle('Analyse Univariée — Variables Numériques', fontsize=16, fontweight='bold', y=1.002)
plt.tight_layout()
plt.savefig('phase1_univariate_numeric.png', bbox_inches='tight', dpi=120)
plt.show()
print('Graphiques enregistrés : phase1_univariate_numeric.png')
```

<h3 style="color: #F39C12; font-weight: bold; font-family: 'Times New Roman';text-align: center;"> Corrélations Pearson  :</h3>

```python
# ── Matrice de corrélation complète
corr_cols = ['stock_initial', 'achat_qte', 'vente_qte', 'stock_final',
             'ecart_stock', 'prix_achat', 'prix_vente',
             'rotation_stock', 'couverture_stock',
             'moyenne_vente_7j', 'moyenne_vente_30j',
             'mois', 'jour_semaine', 'is_weekend',
             'rupture_stock', 'surstock']

corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(16, 13))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # afficher seulement triangle inférieur
sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    fmt='.2f',
    cmap='RdYlGn',
    center=0,
    vmin=-1, vmax=1,
    linewidths=0.5,
    annot_kws={'size': 8},
    square=True
)
plt.title('Matrice de Corrélation de Pearson\n(Triangle inférieur)', fontsize=15, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('phase1_correlation_matrix.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
num_cols = df.select_dtypes(include=np.number).columns

# Fonction pour détecter les outliers par la méthode IQR
def detect_outliers(df, cols):
    outliers = {}
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Identifier les valeurs aberrantes
        outlier_values = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
        outliers[col] = outlier_values.values
    return outliers

# Appliquer la fonction
outliers_dict = detect_outliers(df, num_cols)

# Afficher toutes les valeurs aberrantes et leur nombre
for col, vals in outliers_dict.items():
    print(f"\nColonne: {col}")
    print(f"Nombre d'outliers: {len(vals)}")
    print(f"Valeurs aberrantes: {vals}")
```
